```python
import pandas as pd

dfannealing = pd.read_csv("annealing_proj/dfannealing.csv")
dfthermal = pd.read_csv("annealing_proj/dfthermal.csv")

# Display the first few rows of each DataFrame to verify the import
print("First DataFrame:")
print(dfthermal.head())

print("\nSecond DataFrame:")
print(dfannealing.head())
```

```python
import matplotlib.pyplot as plt
import scienceplots

```

```python
# Plot the   D_KL values
# plt.rc('font', family='serif')
with plt.style.context(['science','ieee']):
    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(1, 1, 1)
    plt.plot(dfannealing['epoch'], dfannealing['D_KL'], label='Annealing')
    plt.plot(dfthermal['epoch'], dfthermal['D_KL'], label='Thermal')
    plt.xlabel('Iteration')
    plt.ylabel('$D_KL(P_v^{data}||P_v)$')
    plt.title('bQRBM vs Annealing based bQRBM')
    # soft grid
    ax.grid(which='major', linestyle='--', linewidth=0.5)
    #ax.grid(which='minor', linestyle=':', linewidth=0.5)
    plt.legend()
    # save as svg
    plt.savefig('annealingVSthermal.pdf', format='pdf', bbox_inches='tight')
    plt.show()
```

```python
# Plot the epoch_duration values
with plt.style.context(['science']):
    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(1, 1, 1)
    plt.plot(dfannealing['epoch'], dfannealing['epoch_duration'], label='Annealing')
    plt.plot(dfthermal['epoch'], dfthermal['epoch_duration'], label='Thermal')
    plt.xlabel('Iteration')
    plt.ylabel('Epoch Duration')
    plt.title('Epoch Duration: Annealing vs Thermal')
    ax.grid(which='major', linestyle='--', linewidth=0.5)
    plt.legend()
    # save as svg
    plt.savefig('epochDuration.pdf', format='pdf', bbox_inches='tight')
    plt.show()
```

```python
import numpy as np

# Plot the logarithm of D_KL values
with plt.style.context(['science']):
    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(1, 1, 1)
    plt.plot(dfannealing['epoch'], np.log(dfannealing['D_KL']), label='Annealing')
    plt.plot(dfthermal['epoch'], np.log(dfthermal['D_KL']), label='Thermal')
    plt.xlabel('Iteration')
    plt.ylabel('Log($D_{KL}$)')
    plt.title('Logarithm of $D_{KL}$: Annealing vs Thermal')
    ax.grid(which='major', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()
```

```python
dftempannealing = pd.read_csv("annealing_proj/tempannealing.csv", header=None)
dftempthermal = pd.read_csv("annealing_proj/tempthermal.csv", header=None)

# Display the first few rows of each DataFrame to verify the import
print("Temp Annealing DataFrame:")
print(dftempannealing.head())

print("\nTemp Thermal DataFrame:")
print(dftempthermal.head())
```

```python
dftempannealing_reduced = dftempannealing.iloc[::2].reset_index(drop=True)
```

```python
# Plot   data for both temp DataFrames
with plt.style.context(['science', 'ieee']):
    fig, ax = plt.subplots(figsize=(7, 5))
    ax.plot(dftempthermal[0], dftempthermal[1], label='Temp Thermal')
    ax.plot(dftempannealing_reduced[0], dftempannealing_reduced[1], label='Temp Annealing')
    
    ax.set_xlabel('Iteration')
    ax.set_ylabel('$Temp[mK]$')
    ax.set_title('Comparison of Temp Annealing and Temp Thermal')
    ax.grid(which='major', linestyle='--', linewidth=0.5)
    ax.legend()
    # save svg
    plt.savefig('tempComparison.pdf', format='pdf', bbox_inches='tight')
    plt.show()
```

```python
# Load A.csv and B.csv
dfA = pd.read_csv("annealing_proj/A.csv", header=None)
dfB = pd.read_csv("annealing_proj/B.csv", header=None)
```
